<h1>Data Cleaning - Ecommerce Platform</h1>

<h4>Step 1: Completeness</h4>
E.g. Handle Missing Values.

<h4>Step 2: Tidiness</h4>
Fix messy structure. --> E.g Split Full Name / Melt-Pivot Data / Normalize Tables

<h4>Step 3: Validity</h4>
Fix impossible values. --> E.g. Age = -5 / Salary = -10000 / Future DOB

<h4>Step 4: Accuracy</h4>
Fix suspicious values. --> E.g. Weight = 1 kg / Age = 250

<h4>Step 5: Consistency</h4>
Standardize values. --> E.g. Male / MALE / male / M
