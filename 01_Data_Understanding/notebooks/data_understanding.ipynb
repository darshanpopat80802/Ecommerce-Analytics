<h1>Data Understanding - Ecommerce Platform</h1>

<h2>Summary for Your Data</h2>

This enterprise-grade e-commerce dataset simulates the operations of a modern online retail business, covering customers, products, orders, payments, inventory, warehouses, suppliers, promotions, shipments, reviews, and employee management. The database consists of 20 interconnected tables designed to support end-to-end business processes and analytical workflows.

<h2>Table Wise Column Description</h2>

**Table -> customers:**

• customer_id: the unique identifier assigned to each customer registered on the e-commerce platform.

• first_name: the given name (i.e. first name) of the customer provided during account registration.

• last_name: the surname (i.e. last name) of the customer used for identification and account management purposes.

• email: the primary email address associated with each customer account, used for login authentication, order confirmations, promotional communications, and customer support interactions.

• phone: the primary contact number provided by the customer for communication regarding orders, deliveries, and account-related activities.

• gender: the gender specified by the customer during account creation, used for customer demographics and marketing analysis.

• birth_date: the date of birth of each customer (month/day/year), used for age segmentation, customer profiling, and personalized marketing campaigns.

• registration_date: the date on which the customer successfully registered an account on the e-commerce platform.

• loyalty_points: the cumulative reward points earned by the customer through purchases, referrals, promotions, and loyalty programs.

---
**Table -> customer_addresses:**

• address_id: the unique identifier assigned to each customer address record stored within the e-commerce platform.

• customer_id: the identifier of the customer associated with the address record. This field links the address to a specific customer account in the customers table.

• address_type: the classification of the address (e.g. billing address, shipping address, home address, or office address).

• address_line_1: the primary street address information including house number, building name, street name, or society name.

• address_line_2: additional address information such as apartment number, floor number, suite number, landmark, or locality details.

• city: the city corresponding to the customer's address.

• state: the state or province corresponding to the customer's address.

• country: the country corresponding to the customer's address.

• postal_code: the postal code or ZIP code associated with the customer's address, used for delivery routing and geographic identification.

---
**Table -> categories:**

• category_id: the unique identifier assigned to each product category available within the e-commerce catalog.

• category_name: the descriptive name used to identify and classify products into a specific category.

• parent_category_id: the identifier of the parent category when the category belongs to a hierarchical category structure. This field allows the creation of subcategories and multi-level product classifications.

---
**Table -> brands:**

• brand_id: the unique identifier assigned to each product brand available on the e-commerce platform.

• brand_name: the official name of the brand associated with one or more products sold through the platform.

• country: the country in which the brand is headquartered or primarily operates.

---
**Table -> suppliers:**

• supplier_id: the unique identifier assigned to each supplier responsible for providing products to the e-commerce business.

• supplier_name: the official business name of the supplier organization supplying products to the company.

• contact_person: the designated representative from the supplier organization who serves as the primary business contact.

• email: the official business email address used for supplier communications, procurement activities, and inventory management coordination.

• phone: the contact number of the supplier or supplier representative used for operational communication.

• city: the city in which the supplier's primary business office is located.

• state: the state or province in which the supplier operates.

• country: the country where the supplier organization is headquartered or conducts business operations.

**Table -> products:**

• product_id: the unique identifier assigned to each product available for sale on the e-commerce platform.

• category_id: the identifier of the category to which the product belongs. This field links the product to a specific category in the categories table.

• brand_id: the identifier of the brand associated with the product. This field links the product to a specific brand in the brands table.

• supplier_id: the identifier of the supplier responsible for providing the product. This field links the product to a supplier in the suppliers table.

• product_name: the official name or title of the product displayed on the e-commerce platform.

• description: the detailed description of the product including its features, specifications, usage information, and selling points.

• cost_price: the amount paid by the company to acquire the product from the supplier before any markup is applied.

• selling_price: the retail price at which the product is offered to customers on the e-commerce platform.

• launch_date: the date on which the product was introduced or made available for sale.

• active_status: the current availability status of the product indicating whether the product is active and available for purchase.

---
**Table -> warehouses:**

• warehouse_id: the unique identifier assigned to each warehouse facility used for storing inventory.

• warehouse_name: the official name used to identify a warehouse location within the company's logistics network.

• city: the city in which the warehouse is physically located.

• state: the state or province in which the warehouse operates.

• country: the country where the warehouse facility is located.

---
**Table -> inventory:**

• inventory_id: the unique identifier assigned to each inventory record maintained within the warehouse management system.

• product_id: the identifier of the product associated with the inventory record. This field links the inventory entry to a specific product in the products table.

• warehouse_id: the identifier of the warehouse where the product inventory is stored. This field links the inventory record to a specific warehouse.

• quantity_available: the current number of units available in stock for the product at the specified warehouse location.

• reorder_level: the minimum stock threshold established for the product. When inventory falls below this level, replenishment or procurement activities may be initiated.

• last_updated: the timestamp indicating when the inventory record was last modified or synchronized within the inventory management system.

---
**Table -> departments:**

• department_id: the unique identifier assigned to each department within the organization.

• department_name: the official name of the department responsible for a specific business function such as Sales, Operations, Marketing, Human Resources, Finance, or Customer Support.

---
**Table -> employees:***

• employee_id: the unique identifier assigned to each employee working within the organization.

• department_id: the identifier of the department to which the employee belongs. This field links the employee to a specific department in the departments table.

• first_name: the given name (i.e. first name) of the employee.

• last_name: the surname (i.e. last name) of the employee.

• email: the official company email address assigned to the employee for business communication and internal operations.

• hire_date: the date on which the employee officially joined the organization.

• salary: the annual compensation amount paid to the employee by the organization, used for payroll administration and workforce analysis.

**Table -> orders:**

• order_id: the unique identifier assigned to each customer order placed on the e-commerce platform.

• customer_id: the identifier of the customer who placed the order. This field links the order to a specific customer in the customers table.

• order_date: the date and time at which the order was successfully placed by the customer.

• order_status: the current status of the order within the order fulfillment lifecycle (e.g. Pending, Confirmed, Packed, Shipped, Delivered, Cancelled, or Returned).

• total_amount: the total monetary value of the order including all purchased products before refunds are applied.

---
**Table -> order_items:**

• order_item_id: the unique identifier assigned to each product line item within an order.

• order_id: the identifier of the order associated with the product line item. This field links the item to a specific order in the orders table.

• product_id: the identifier of the product purchased within the order. This field links the order item to a specific product in the products table.

• quantity: the number of units of the product purchased by the customer within the order.

• unit_price: the selling price of a single unit of the product at the time the order was placed.

• subtotal: the total monetary value for the line item calculated based on the product quantity and unit price.

---
**Table -> payments:**

• payment_id: the unique identifier assigned to each payment transaction recorded within the platform.

• order_id: the identifier of the order associated with the payment. This field links the payment to a specific order in the orders table.

• payment_date: the date and time at which the payment transaction was processed.

• payment_method: the method used by the customer to complete the payment (e.g. Credit Card, Debit Card, UPI, Net Banking, Wallet, or Cash on Delivery).

• payment_status: the current status of the payment transaction (e.g. Pending, Successful, Failed, Refunded, or Cancelled).

• amount: the total monetary amount processed during the payment transaction.

---
**Table -> shipments:**

• shipment_id: the unique identifier assigned to each shipment created for customer order fulfillment.

• order_id: the identifier of the order associated with the shipment. This field links the shipment to a specific order in the orders table.

• warehouse_id: the identifier of the warehouse responsible for dispatching the products included in the shipment.

• shipped_date: the date and time at which the shipment was dispatched from the warehouse.

• delivery_date: the date and time at which the shipment was successfully delivered to the customer.

• shipping_cost: the total logistics cost incurred for transporting the shipment from the warehouse to the customer.

• tracking_number: the unique tracking reference assigned by the logistics provider to monitor shipment progress and delivery status.

---
**Table -> returns:**

• return_id: the unique identifier assigned to each product return request submitted by a customer.

• order_id: the identifier of the order associated with the returned product. This field links the return record to a specific order in the orders table.

• product_id: the identifier of the product being returned by the customer. This field links the return request to a specific product in the products table.

• return_date: the date on which the customer initiated or completed the return request.

• reason: the explanation provided for the product return (e.g. Damaged Product, Wrong Item Delivered, Defective Product, Size Issue, or Customer Preference).

• refund_amount: the monetary amount refunded to the customer as part of the return process.

**Table -> reviews:**

• review_id: the unique identifier assigned to each customer review submitted on the e-commerce platform.

• customer_id: the identifier of the customer who submitted the review. This field links the review to a specific customer in the customers table.

• product_id: the identifier of the product being reviewed. This field links the review to a specific product in the products table.

• rating: the numerical score assigned by the customer to evaluate the product, typically measured on a predefined rating scale (e.g. 1 to 5 stars).

• review_text: the written feedback provided by the customer describing their experience, satisfaction level, product quality, and recommendations.

• review_date: the date and time at which the review was submitted by the customer.

---
**Table -> promotions:**

• promotion_id: the unique identifier assigned to each promotional campaign created by the business.

• promotion_name: the descriptive name used to identify and manage the promotional campaign.

• start_date: the date on which the promotional campaign becomes active and available to customers.

• end_date: the date on which the promotional campaign expires and is no longer applicable.

• discount_percent: the percentage discount offered to customers as part of the promotion.

---
**Table -> product_promotions:**

• product_id: the identifier of the product associated with a promotional campaign. This field links the record to a specific product in the products table.

• promotion_id: the identifier of the promotion applied to the product. This field links the record to a specific promotion in the promotions table.

Note: This table acts as a junction (bridge) table that enables a many-to-many relationship between products and promotions. A product can participate in multiple promotions, and a promotion can be applied to multiple products.

---
**Table -> order_management:**

• management_id: the unique identifier assigned to each order management activity performed within the organization.

• order_id: the identifier of the order associated with the management activity. This field links the activity to a specific order in the orders table.

• employee_id: the identifier of the employee responsible for performing the management action. This field links the activity to a specific employee in the employees table.

• action_date: the date and time at which the management action was performed.

• action_type: the type of operational activity performed on the order (e.g. Order Confirmed, Order Packed, Quality Checked, Shipment Assigned, Delivery Verified, or Return Processed).

---
**Table -> sales_targets:**

• target_id: the unique identifier assigned to each sales target record established by the organization.

• employee_id: the identifier of the employee assigned the sales target. This field links the target to a specific employee in the employees table.

• target_month: the calendar month for which the sales target has been assigned.

• target_year: the calendar year for which the sales target has been assigned.

• target_amount: the monetary sales objective assigned to the employee for the specified month and year.
