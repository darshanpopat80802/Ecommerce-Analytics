# Customer Data Exploratory Data Analysis (EDA)

## Project Overview

This notebook performs exploratory data analysis on the Customers table of the Enterprise E-Commerce Database.

The objective is to understand customer demographics, identify data quality issues, analyze customer registration trends, study loyalty behavior, detect outliers, and generate actionable business insights that can support downstream machine learning models such as Customer Churn Prediction and Customer Lifetime Value (CLV) Prediction.

```python
df["registration_date"] = pd.to_datetime(df["registration_date"])

df["birth_date"] = pd.to_datetime(df["birth_date"])
```

11. Loyalty Points Analysis

```python
df["loyalty_points"].describe()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["loyalty_points"],bins=30)

plt.xlabel("Loyalty Points")

plt.ylabel("Customers")

plt.title("Distribution of Loyalty Points")

plt.show()
```

```python
plt.figure(figsize=(8,2))

sns.boxplot(x=df["loyalty_points"])

plt.show()
```

12. Create Age Feature

```python
today = pd.Timestamp.today()

df["age"] = (
    today - df["birth_date"]
).dt.days//365
```

13. Age Distribution

```python
df["age"].describe()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["age"],bins=20)

plt.xlabel("Age")

plt.ylabel("Customers")

plt.title("Customer Age Distribution")

plt.show()
```

```python
sns.boxplot(x=df["age"])

plt.show()
```

14. Create Age Groups

```python
bins=[18,25,35,45,60,100]

labels=[
    "18-25",
    "26-35",
    "36-45",
    "46-60",
    "60+"
]

df["age_group"]=pd.cut(
    df["age"],
    bins=bins,
    labels=labels
)
```

```python
sns.countplot(data=df,x="age_group")

plt.show()
```

15. Registration Trend

```python
df["registration_year"]=df["registration_date"].dt.year
```

```python
df["registration_month"]=df["registration_date"].dt.month_name()
```

```python
df.groupby("registration_year").size()
```

```python
df.groupby("registration_year").size().plot(
    marker="o",
    figsize=(8,5)
)

plt.ylabel("Customers")

plt.show()
```

```python
order=[
"January",
"February",
"March",
"April",
"May",
"June",
"July",
"August",
"September",
"October",
"November",
"December"
]

sns.countplot(
    data=df,
    x="registration_month",
    order=order
)

plt.xticks(rotation=45)

plt.show()
```

16. Customer Segmentation

```python
bins=[0,100,500,1000,5000]

labels=[
    "Low",
    "Medium",
    "High",
    "VIP"
]

df["customer_segment"]=pd.cut(
    df["loyalty_points"],
    bins=bins,
    labels=labels
)
```

```python
sns.countplot(
    data=df,
    x="customer_segment"
)

plt.show()
```

17. Correlation Analysis

```python
numeric=df.select_dtypes(include=np.number)

numeric.corr()
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    numeric.corr(),
    annot=True,
    cmap="coolwarm"
)

plt.show()
```

18. Key Performance Indicators (KPIs)

```python
print("Total Customers :",len(df))

print("Average Age :",round(df["age"].mean(),2))

print("Average Loyalty Points :",round(df["loyalty_points"].mean(),2))

print("Median Loyalty Points :",df["loyalty_points"].median())

print("Maximum Loyalty Points :",df["loyalty_points"].max())
```
